```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from pandas.plotting import table
from scipy.stats import pareto, norm
```

```python
# Generate power law distributed data
data = np.random.pareto(2, 1000)

# Calculate standard deviation and mean
std_dev = np.std(data)
mean = np.mean(data)

# Define the probability density function for the power law distribution
x = np.linspace(pareto.ppf(0.01, 2), pareto.ppf(0.99, 2), 100)
pdf = pareto.pdf(x, 2)

# panda dataframe
df = pd.DataFrame(data)

# Plot the data
fig, ax = plt.subplots()
ax.hist(data, bins=50, density=True, alpha=0.7)
ax.plot(x, pdf, color='black', linestyle='--', label='PDF')
table(ax, np.round(df.describe(), 2), loc="upper right", colWidths=[0.2, 0.2, 0.2])
ax.axvline(x=mean, color='red', linestyle='--', label='Mean')
ax.axvline(x=mean + std_dev, color='orange', linestyle='--', label='1 Std Dev')
ax.axvline(x=mean - std_dev, color='orange', linestyle='--')
ax.set_xlabel('Data Values')
ax.set_ylabel('Frequency / PDF')
ax.set_title('Power Law Distributed Data')
plt.show()
```

This code generates 1000 random values that follow a power law distribution with a shape parameter of 2 using the numpy.random.pareto() function. 

Then, it creates a histogram plot of the data using the matplotlib.pyplot.hist() function, with 50 bins, normalized density values, and an opacity of 0.7. 

The resulting plot shows a long tail on the right-hand side of the histogram, indicating that the data follows a power law distribution.

$$f(x) = \frac{\alpha - 1}{x^\alpha}$$
This is known as the "power law density function" or the "Pareto distribution", which is the continuous version of the discrete power law distribution. 

The Pareto distribution has a number of interesting properties and is widely used in economics, finance, and other fields to model phenomena with *heavy-tailed distributions*.

```python
# Generate normally distributed data
data = np.random.normal(0, 1, 1000)

# Calculate standard deviation and mean
std_dev = np.std(data)
mean = np.mean(data)

# Define the probability density function for the normal distribution
x = np.linspace(norm.ppf(0.01), norm.ppf(0.99), 100)
pdf = norm.pdf(x)

# panda dataframe
df = pd.DataFrame(data)

# Plot the data
fig, ax = plt.subplots()
ax.hist(data, bins=50, density=True, alpha=0.7)
ax.plot(x, pdf, color='black', linestyle='--', label='PDF')
table(ax, np.round(df.describe(), 2), loc="upper right", colWidths=[0.2, 0.2, 0.2])
ax.axvline(x=mean, color='red', linestyle='--', label='Mean')
ax.axvline(x=mean + std_dev, color='orange', linestyle='--', label='1 Std Dev')
ax.axvline(x=mean - std_dev, color='orange', linestyle='--')
ax.set_xlabel('Data Values')
ax.set_ylabel('Frequency / PDF')
ax.set_title('Normally Distributed Data')
plt.show()
```

This code generates 1000 random values that follow a normal distribution with a mean of 0 and a standard deviation of 1 using the numpy.random.normal() function.

Then, it creates a histogram plot of the data using the matplotlib.pyplot.hist() function, with 50 bins, normalized density values, and an opacity of 0.7. 

The resulting plot shows a symmetric bell curve shape in the histogram, indicating that the data follows a normal distribution.

### Sampling
Here's an example of creating a sample normal distribution tensor and using TensorFlow sampling with random selection with replacement in Python:

```python
import tensorflow as tf
import numpy as np

# Create a sample normal distribution tensor with mean 0 and standard deviation 1
data = tf.random.normal([1000], mean=0, stddev=1)

def sample_data(data, n_samples):
  # Use TensorFlow sampling to randomly select n_samples data points with replacement
  sampled_data = tf.random.shuffle(data)[:n_samples]
  return sampled_data.numpy()

# Sample 10 data points from the normal distribution tensor
sampled_data = sample_data(data, 10)
print(sampled_data)
```
